```
import pandas as pd
import matplotlib.pyplot as plt
```

```
alphabet = pd.read_csv('alphabet_stock_data.csv')
alphabet.head(2)
```

```
alphabet['Date'] = pd.to_datetime(alphabet.Date, format='%Y-%m-%d')
alphabet.info()
```

```
alphabet.tail()
```

```
data = alphabet[("2020-04-01 00:00:00" <= alphabet['Date']) & (alphabet['Date'] <= "2020-09-24 00:00:00")]
```

```

plt.plot(data.Date, data.Close)
```
